```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Data Preprocessing

```python
df.shape
```

```python
#check data type of columns
df.dtypes
```

temperatue should be object it should be either integer or float


```python
df['Temperatue'].unique()
```

need to remove : and convert it to float type

```python
df=df[df['Temperatue']!=":"]
```

```python
df['Temperatue']=df['Temperatue'].astype(float)
```

```python
#check for missing values
df.isnull().sum()
```

```python
#replace missing values with median
columns=[df.columns]
for col in columns:
  df[col]=df[col].fillna(df[col].median())
```

Train Test split

```python
from sklearn.model_selection import train_test_split
x=df.drop('Yeild (Q/acre)',axis=1)
y=df['Yeild (Q/acre)']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

##Model Building
1.Decision tree regressor
2. random forest regressor

Decision tree regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
dtr=DecisionTreeRegressor()
```

hyperparameter tuning using gridsearch

```python
from sklearn.model_selection import GridSearchCV
para={
    'max_depth':[2,4,6,8,10],
    'min_samples_split':[2,4,6,8,10],
    'min_samples_leaf':[2,4,6,8,10],
    'random_state':[0,42]
}
grid=GridSearchCV(dtr,param_grid=para,cv=5)
grid.fit(x_train,y_train)

print(grid.best_params_)
```

```python
#decision tree regressor with best paraameters
dtr=DecisionTreeRegressor(max_depth=4,min_samples_leaf=2,min_samples_split=10,random_state=0)

dtr.fit(x_train,y_train)

print(dtr.score(x_train,y_train))

d_pred=dtr.predict(x_test)

print(r2_score(y_test,d_pred))
```

##random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
```

hyperparameter tuning using gridsearchcv

```python
from sklearn.model_selection import GridSearchCV
para={
    'max_depth':[2,4,6,8,10],
    'min_samples_split':[2,4,6,8,10],
    'min_samples_leaf':[2,4,6,8,10],
    'random_state':[0,42]
}
grid=GridSearchCV(rfr,param_grid=para,cv=5)
grid.fit(x_train,y_train)

print(grid.best_params_)
```

```python
rfr=RandomForestRegressor(max_depth=4,min_samples_leaf=2,min_samples_split=6,n_estimators=100,random_state=42)
rfr.fit(x_train,y_train)
print(rfr.score(x_train,y_train))
r_pred=rfr.predict(x_test)
print(r2_score(y_test,r_pred))
```

##Model evaluation

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))
#actual vs predicted for decison tree
sns.distplot(y_test,ax=ax[0]).set_title('Actual')
sns.distplot(d_pred,ax=ax[0]).set_title('Decision tree regressor')

#actual vs predicted for random forest
sns.distplot(y_test,ax=ax[1]).set_title('Actual')
sns.distplot(r_pred,ax=ax[1]).set_title('random forest regressor')
plt.show()
```

##Evaluation metrics

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print('decision tree regressor')
print('mean absolute error:',mean_absolute_error(y_test,d_pred))
print('mean squared error:',mean_squared_error(y_test,d_pred))
print('r_2 score',r2_score(y_test,d_pred))
print('\n')
print('random forest regresor')
print('mean absolute error:',mean_absolute_error(y_test,r_pred))
print('mean squared error:',mean_squared_error(y_test,r_pred))
print('r_2 score',r2_score(y_test,r_pred))
```

##Feature importance

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))
fig.subplots_adjust(wspace=0.5)

#decision tree regressor
feature_df = pd.DataFrame({'Features':x_train.columns, 'Importance':dtr.feature_importances_})
feature_df.sort_values(by='Importance', ascending=False, inplace=True)
sns.barplot(x = 'Importance', y = 'Features', data = feature_df, ax = ax[0]).set_title('Decision Tree Regressor')

#random forest regresor
feature_df = pd.DataFrame({'Features':x_train.columns, 'Importance':rfr.feature_importances_})
feature_df.sort_values(by='Importance', ascending=False, inplace=True)
sns.barplot(x = 'Importance', y = 'Features', data = feature_df, ax = ax[1]).set_title('Random Forest Regressor')
```
